```python
"""reading data into single variable 'dataset'"""

dataset = pd.read_csv('DigitalAd_dataset.csv')
print(dataset)
```

```python
"""summarizing dataset"""

print(dataset.shape)
print(dataset.head(5))
```

```python
"""segrigating data into input(i.e. x) and output (i.e. y)"""

x=dataset.iloc[:,:-1].values
x

y=dataset.iloc[:,-1].values
y
```

```python
"""splitting the dataset into train and test"""

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
"""preprocessing data through feature scaling"""

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import *
model = LogisticRegression()
model.fit(x_train, y_train)

age=int(input("Enter New Customer's age:"))
sal=int(input("Enter New Customer's salary:"))
newCust=[[age,sal]]
result=model.predict(sc.transform(newCust))
print(result)
if result==1:
  print("customer will buy")
else:
  print("customer will not buy")

y_pred=model.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

from sklearn.metrics import confusion_matrix, accuracy_score
cm= confusion_matrix(y_test, y_pred)

print("confusion matrix:" )
print(cm)

print("accuracy of the model: {0}%".format(accuracy_score(y_test, y_pred)*100))
```
